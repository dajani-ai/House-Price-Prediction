```python
# Import libraries
import pandas as pd
```

```python
df = pd.read_csv('housing.csv')
```

```python
# Handling logical inconcistency
df = df[df['households'] <= df['population']]
```

```python
# Handling null values 
df.dropna(inplace= True)
```

```python
df.to_csv('housing_cleaned.csv', index=False)
```
